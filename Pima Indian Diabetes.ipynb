## Pima Indians Diabetes Dataset

### Content
The datasets consists of several medical predictor variables and one target variable, Outcome. Predictor variables includes the number of pregnancies the patient has had, their BMI, insulin level, age, and so on

### Data Dictionary 
Pregnancies-
Number of times pregnant

Glucose-
Plasma glucose concentration a 2 hours in an oral glucose tolerance test

BloodPressure-
Diastolic blood pressure (mm Hg)

SkinThickness-
Triceps skin fold thickness (mm)

Insulin-
2-Hour serum insulin (mu U/ml)

BMI-
Body mass index (weight in kg/(height in m)^2)

DiabetesPedigreeFunction-
Diabetes pedigree function

Age-
Age (years)

Outcome-
Class variable (0 or 1) 268 of 768 are 1, the others are 0

# Data Cleaning

## Removing Outliers


### For Checking the 'Pregnancies' column 

```python
new_data['Pregnancies'].value_counts()
```

```python
sns.distplot(new_data['Pregnancies'], bins=10)
plt.title('Histogram')
plt.show()
```

```python
stats.probplot(new_data['Pregnancies'], dist='norm', fit=True, plot=plt)
plt.title('Q-Q Plot')
```

```python
# taking 95% percentile
max_threshold = new_data['Pregnancies'].quantile(0.95)
new_data = new_data[new_data['Pregnancies']<=max_threshold]
```

### Checking the 'Glucose' column 

```python
sns.distplot(new_data['Glucose'], bins=10)
plt.title('Histogram')
plt.show()
```

```python
 stats.probplot(new_data['Glucose'], dist='norm', fit=True, plot=plt)
plt.title('Q-Q Plot')
```

```python
new_data = new_data[new_data['Glucose']>=25]   # removal of outliers as can be seen from above graphs
```

### Checking the 'BloodPressure' column -

```python
sns.distplot(new_data['BloodPressure'], bins=10)
plt.title('Histogram')
plt.show()
```

```python
 stats.probplot(new_data['BloodPressure'], dist='norm', fit=True, plot=plt)
plt.title('Q-Q Plot')
```

```python
new_data = new_data[new_data['BloodPressure']!=0]   # removal of outliers as can be seen from above graphs
```

### Checking the 'BMI' column -

```python
sns.distplot(new_data['BMI'], bins=10)
plt.title('Histogram')
plt.show()
```

```python
stats.probplot(new_data['BMI'], dist='norm', fit=True, plot=plt)
plt.title('Q-Q Plot')
```

```python
new_data = new_data[(new_data['BMI']>10) & (new_data['BMI']<50)]   # removal of outliers as can be seen from above graphs
```

### Checking the 'Insulin' column -

```python
sns.distplot(new_data['Insulin'], bins=10)
plt.title('Histogram')
plt.show()
```

```python
stats.probplot(new_data['Insulin'], dist='norm', fit=True, plot=plt)
plt.title('Q-Q Plot')
```

```python
# we should check columns
new_data['Insulin'].value_counts().sort_index(ascending=False)
```

```python
new_data = new_data[new_data['Insulin'].between(15, 600)] # removing outliers
```

### Checking the 'SkinThickness' column 

```python
sns.distplot(new_data['SkinThickness'], bins=10)
plt.title('Histogram')
plt.show()
```

```python
stats.probplot(new_data['SkinThickness'], dist='norm', fit=True, plot=plt)
plt.title('Q-Q Plot')
```

```python
new_data = new_data[(new_data['BMI']<60)] # removing outliers
```

### Checking the 'DiabetesPedigreeFunction' column -

```python
sns.distplot(new_data['DiabetesPedigreeFunction'], bins=10)
plt.title('Histogram')
plt.show()
```

```python
stats.probplot(new_data['DiabetesPedigreeFunction'], dist='norm', fit=True, plot=plt)
plt.title('Q-Q Plot')
```

```python
new_data = new_data[(new_data['DiabetesPedigreeFunction']<1.5)] # removing outliers
```
